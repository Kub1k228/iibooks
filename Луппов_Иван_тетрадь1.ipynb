# 1.1. Теоретический материал – Типы данных
Типы данных
Все типы данных в Python относятся к одной из 2-х категорий: изменяемые (mutable) и неизменяемые (immutable).
Неизменяемые объекты:
- числовые данные (int, float),
- bool,
- None,
- символьные строки (class 'str'),
- кортежи (tuple).
Изменяемые объекты:
- списки (list),
- множества (set),
- словари (dict).

```python
#Пример 1.2
x = 3+2.5*7
y = None
z = 'a', 5, 12.345, (2,'b')
df = [['Антонова Антонина', 34, 'ж'], ['Борисов Борис',26,'M']]
A = {1,'tile',2,'content'}
print (x,'|',type(x),'\n',y,'|',type(y),'\n',z,'|',type(z),'\n',df,'|',type(df),'\n',A,'|',type(A))
```

# 2.1. Теоретический материал – Условный оператор
If – Условный оператор
В коде часто приходится проверять выполнимость или невыполнимость каких-то условий. Синтаксис:
if <условие1 (булевское выражение)> :
<код, который выполнится, если условие верно> elif <условие2 (булевское выражение)>:
<код, который выполнится, если условие1 было неверно, а условие2 верно>
else:
<код, который выполнится, если условие1 и условие2 были неверны>
Обратите внимание, что код, который должен выполняться внутри каждого условия, записывается с отступом в 4 пробела от уровня if, elif и else: в питоне области видимости переменных обозначаются отступами.
То есть, отступы позволяют понять, где начинается код, который должен выполняться при выполнении условия в if, и где заканчивается.

```python
#Пример 2.2
x= 0
if x>0:
    print ("Positiv")
elif (x<0):
    print("Negative")
else:
    print ("zero")
```

```python
#4.2.2 Пример
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
from numpy import trapz
x = np.arange (0.0,10,0.1#)
y = np.abs(np.sin(x*np.exp(np.cos(x))))
plt.grid()
plt.plot(x,y,c = "r")
plt.fill_between(x,y)

area = trapz(y)
print (area)
```

```python
#4.3.2 Задание
from matplotlib import pyplot as plt
import math as mt
import numpy as np
def f(x):
    f1 = 1+(mt.exp(x**0.5))+(mt.cos(x**2))
    f2 = abs(1-(mt.sin(x)**3))
    f3 = mt.log(abs(2*x))
    return (f1/f2)+f3


a = np.empty((10))
for i in range(1,10):
    a[i]=f(i)
x=a[0:5:1]
plt.plot(x,'ro')
plt.plot(a)
plt.show
```

```python
import math

def calculator():
    print("Веберите операцию: +, -, *, /, sin, cos, pow, exp")
    operation = input("Введите операцию: ")
    if operation == "+" or operation == "-" or operation == "*" or operation == "/":
        x = float(input("Введите первое число: "))
        y = float(input("Введите второе число: "))
        if operation == "+":
            print(f"Результат: {x + y}")
        elif operation == "-":
            print(f"Результат: {x - y}")
        elif operation == "*":
            print(f"Результат: {x * y}")
        elif operation == "/":
            if y != 0:
                print(f"Результат: {x / y}")
            else:
                print("А на ноль делить нельзя))")
    elif operation == "sin" or operation == "cos":
        x = float(input("Введите первое число: "))
        y = float(input("Введите второе число: "))
        angle = math.radians(x + y)
        if operation == "sin":
            print(f"Результат: {math.sin(angle)}")
        elif operation == "cos":
            print(f"Результат: {math.cos(angle)}")
    elif operation == "pow" or operation == "exp":
        x = float(input("Введите первое число: "))
        y = float(input("Введите второе число: "))
        if operation == "pow":
            print(f"Результат: {pow(x, y)}")
        elif operation == "exp":
            print(f"Результат: {math.exp(x + y)}")
    else:
        print("Ошибка при вводе опереции!")

calculator()
```
